**Let $T_s=1\text{ sec}$, $f_c=0.45\text{ Hz}$, and $N=20$.**

**1. Compute and plot the magnitude response of the system in Figure 2 without the sample-and-hold device. Let the frequencies be ```f = np.linspace(-1, 1, 2001)```.**

````python
# write your code here
T = 1
f_c = 0.45
N = 20
f = np.linspace(-1,1,2001)
H_b = 1/(1+(f/f_c)**N)

plt.plot(f, H_b)
plt.title("1.1: Magnitude Response without the sample-and-hold device")
plt.xlabel("frequency")
plt.ylabel("magnitude")
````

**2. Compute and plot the magnitude response of the complete system in Figure 2. Let the frequencies be ```f = np.linspace(-1, 1, 2001)```.**

````python
# write your code here
T = 1
f_c = 0.45
N = 20
f = np.linspace(-1,1,2001)
H = (1/(1+(f/f_c)**N))**2 * np.abs(np.sinc(f * T))

plt.plot(f, H)
plt.title("1.2: Magnitude Response of Complete System")
plt.xlabel("frequency")
plt.ylabel("magnitude")
````

**3. Comment on the shape of the two magnitude responses. How might the magnitude response of the sample-and-hold affect the design considerations of a high quality audio CD player?**

The magnitude response is significantly flatter for the system without the sample-and-hold device.  The frequency roll off in the magnitude response of the sample-and-hold will result in higher frequencies being less pronounced than they should be.  Specifically, for this case there would be more muted mids and highs.  When designing a high quality audio CD player, it might be worth adding some sort of digital amplifier/filter (amplify the higher frequencies) to help compensate for the frequency roll off present in the sample-and-hold device.
